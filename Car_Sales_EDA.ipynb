```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Dataset
url = 'https://raw.githubusercontent.com/Shivam-kumar20/Car_Sales_EDA/main/Car_Sales_Raw.csv'

df = pd.read_csv(url)
```

```python
# Checking Dataframe

df.head(5)
```

```python
# Checking Datatype

df.dtypes
```

```python
# # Changing Date Datatype

from datetime import datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
# Checking Duplicated

df[df.duplicated()]
```

```python
df.head(5)
```

```python
# Max and Minimum Sales Price

df.describe()
```

```python
df.info()
```

```python
# Price ($) Converting to INR

df['Price_INR'] = df['Price ($)'].apply(lambda x : x*83.68)
df['Annual Income_INR'] = df['Annual Income'].apply(lambda x : x*83.68)
```

```python
# Dropping Irrevelant columns

df.drop(columns=['Car_id','Dealer_No ','Customer Name'], axis=1, inplace=True)
```

### EDA - Exploratory Data Analysis

```python

plt.figure(figsize=(30,10))
d1 = df.groupby('Company')['Price ($)'].sum().reset_index().sort_values(by ='Price ($)', ascending=True)

ax = sns.barplot(data = d1,x=df['Company'],y = df['Price ($)'],errorbar=None,estimator = 'sum')
plt.xlabel('Companies')
plt.ylabel('Price in Dollars')
for bars in ax.containers:
    ax.bar_label(bars)
    
# Conclusion -> Chevrolet, Ford, Dodge and oldsmobile are most Revenue generating companies 
# where as Volkswagen is most selling car company
```

```python
plt.figure(figsize=(30,10))
ax = sns.countplot(data = df,x=df['Company'])
plt.xlabel('Companies')
plt.ylabel('No. of Cars Sold')
for bars in ax.containers:
    ax.bar_label(bars)
    
# Conclusion -> 
```

```python
plt.figure(figsize=(30,30))
sns.scatterplot(data = df, y = df['Annual Income'], 
                x = df['Price ($)'],
                hue = 'Gender',
                size = df['Annual Income'])
plt.show()
```

```python
sns.boxplot(data = df,x=df['Annual Income_INR'] )

q1 = df['Annual Income_INR'].quantile(0.25)
q3 = df['Annual Income_INR'].quantile(0.75)

IQR = q3-q1

lower_thre = max((q1-(1.5*IQR)),min(df['Annual Income_INR']))

upper_thre = q3+(1.5*IQR)
print(f'The lower thresold value is {round(lower_thre,2)}')
print(f'The lower thresold value is {round(upper_thre,2)}')
```
